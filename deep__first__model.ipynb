```python
import pandas as pd
df = pd.read_csv("auto-mpg.csv")
```

```python
df.head()
```

```python
df=df.drop("car name",axis=1)
```

```python
df.head()
```

```python
df.info()
```

## DATA_PREPROCESSING

```python
import numpy as np
df["horsepower"]=df["horsepower"].replace("?",np.nan)
```

```python
df["horsepower"]=df["horsepower"].astype(float)
```

```python
df["horsepower"]=df["horsepower"].fillna(df["horsepower"].mean())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df2=sc.fit_transform(df)
```

```python
df2=pd.DataFrame(df2,columns=df.columns)
df2.head()
```

```python
x=df2.iloc[:,1:]
y=df2.iloc[:,0]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x.shape
```

##  model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import metrics
```

```python
model1=Sequential()
model1.add(Dense(64,activation="relu",input_shape=(x.shape[1],)))
model1.add(Dense(16,activation="relu"))
model1.add(Dense(16,activation="relu"))
model1.add(Dense(1,activation="linear"))
```

```python
model1.summary()
```

```python
model1.compile(optimizer="adam",loss="mse",metrics=[metrics.R2Score])
```

```python
model1.fit(x_train,y_train,epochs=90,validation_split=.2)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,model1.predict(x_test))
```

```python
model2=Sequential()
model2.add(Dense(64,activation="tanh",input_shape=(x.shape[1],)))
model2.add(Dense(16,activation="tanh"))
model2.add(Dense(1,activation="linear"))
```

```python
model2.summary()
```

```python
model2.compile(optimizer="adam",loss="mse",metrics=[metrics.R2Score])
```

```python
model2.fit(x_train,y_train,epochs=100,validation_split=.2)
```

```python
r2_score(y_test,model2.predict(x_test))
```

```python
model3=Sequential()
model3.add(Dense(64,activation="tanh",input_shape=(x.shape[1],)))
model3.add(Dense(16,activation="relu"))
model3.add(Dense(1,activation="linear"))
```

```python
model3.summary()
```

```python
model3.compile(optimizer="adam",loss="mse",metrics=[metrics.R2Score])
```

```python
model3.fit(x_train,y_train,epochs=100,validation_split=.2)
```

```python
r2_score(y_test,model3.predict(x_test))
```
